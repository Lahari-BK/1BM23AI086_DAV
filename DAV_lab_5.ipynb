```python
import pandas as pd

df = pd.read_csv('sales_dataset.csv', parse_dates=['Date'])

# Question 1: Total revenue by salesperson and date

q1 = df.pivot_table(index='Salesperson', columns='Date', values='Revenue', aggfunc='sum')
print("Q1: Total Revenue by Salesperson and Date\n", q1)

# Question 2: Average revenue per sale for each product

q2 = df.groupby('Product')['Revenue'].mean()
print("\nQ2: Average Revenue per Sale by Product\n", q2)

# Question 3: Max units sold in a single transaction by each salesperson

q3 = df.groupby('Salesperson')['Units Sold'].max()
print("\nQ3: Max Units Sold per Transaction by Salesperson\n", q3)

# Question 4: Percentage of total revenue by region

total_revenue = df['Revenue'].sum()
q4 = df.groupby('Region')['Revenue'].sum() / total_revenue * 100
print("\nQ4: Percentage of Total Revenue by Region\n", q4)

# Question 5: Salesperson with most transactions

q5 = df['Salesperson'].value_counts()
print("\nQ5: Number of Transactions per Salesperson\n", q5)

# Question 6: Pivot showing total revenue and units sold by salesperson and product

q6 = df.pivot_table(index='Salesperson', columns='Product', values=['Revenue', 'Units Sold'], aggfunc='sum')
print("\nQ6: Total Revenue and Units Sold by Salesperson and Product\n", q6)

# Question 7: Total units sold in each region on each date

q7 = df.pivot_table(index='Region', columns='Date', values='Units Sold', aggfunc='sum')
print("\nQ7: Total Units Sold by Region and Date\n", q7)
```
