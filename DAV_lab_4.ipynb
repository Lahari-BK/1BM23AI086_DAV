```python
import pandas as pd
import numpy as np
df = pd.read_csv('book2.csv')
print("Original Dataset:\n", df, "\n")
print("Pandas mean GDP:", df['gdp'].mean())
print("Pandas sum of Population:", df['population'].sum())
print("Pandas max Life Expectancy:", df['life_expectency'].max())
print("Pandas min Life Expectancy:", df['life_expectency'].min())

print("\nNumPy mean Population:", np.mean(df['population']))
print("NumPy std GDP:", np.std(df['gdp']))
print("NumPy sum GDP:", np.sum(df['gdp']), "\n")
df['GDP_10%_Increase'] = df['gdp'] * 1.10
print("After increasing GDP by 10%:\n", df[['country', 'year', 'gdp', 'GDP_10%_Increase']], "\n")
df_multi = df.set_index(['country', 'year'])
print("MultiIndex DataFrame:\n", df_multi, "\n")

swapped = df_multi.swaplevel().sort_index(level='year')
print("After swapping index levels and sorting by Year:\n", swapped, "\n")
unstacked = df_multi[['gdp', 'life_expectency']].unstack(level='year')
print("Unstacked Data (Year as columns):\n", unstacked, "\n")
print("Population trend for China over years:\n", df_multi.loc['china']['population'], "\n")
print("Population trend for USA over years:\n", df_multi.loc['USA']['population'], "\n")
```
